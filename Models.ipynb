```python
#DATA PROCESSING
#load the dataset
import pandas as pd
```

```python
df = pd.read_csv("Car Dataset Processed.csv")
```

```python
df.head()
```

```python
#to convert text data of columns to numbers
df["insurance_validity"].value_counts()
```

```python
mapping = {'Comprehensive':5,'Third Party insurance':1, 'Zero Dep':2,
                            'Not Available':3,'Third Party':1
          }

df["insurance_validity"]= df['insurance_validity'].map(mapping)
                                                        
```

```python
df["insurance_validity"].value_counts()
```

```python
#now for fuel_type
df["fuel_type"].value_counts()
```

```python
mapping1 = {
            'Petrol':0,
            'Diesel':1,
            'CNG':2
        }
df["fuel_type"]=df["fuel_type"].map(mapping1)
```

```python
df["fuel_type"].value_counts()
```

```python
#now for transmission type
df["transmission"].value_counts()
```

```python
mapping2 = {
            "Manual":5,
            "Automatic":1,
            }
df["transmission"]= df["transmission"].map(mapping2)
```

```python
df["transmission"].value_counts()
```

```python
#for ownership column
df["ownsership"].value_counts()
```

```python
mapping3={
            "First Owner":1,
            "Second Owner":2,
            "Third Owner":3,
            "Fifth Owner":5,
}
df["ownsership"]= df["ownsership"].map(mapping3)
```

```python
df["ownsership"].value_counts()
```

```python
#work on registration year
#make another column wherein there is only the years
df['registration_year'] = df['registration_year'].str.split('-').str[1]
```

```python
df.head()
```

```python
#model development
#insurance_validity,fuel_type,ownsership,kms_driven, transmission
#based on the baove columns we are gonna machine learning model
#as a datascientist these many columns shoule be less we used to use more


```

```python
#Phase 1:input features
#x:insuarance_validity 2d object;all about numeric data
#y:price,

#types of objects:dictonaries
#this is a regression problem

X = df[["insurance_validity", "fuel_type", "kms_driven", "ownsership", "transmission"]]
Y = df["price(in lakhs)"]

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

```

```python
#define models
models= {'LinReg':LinearRegression(),
        'KNN-3':KNeighborsRegressor(n_neighbors=3),
        'SVM-Poly':SVR (kernel='poly')
     }
#trainin and evaluation
for name,model in models.items():
        model.fit(X,Y)      
        score = model.score(X,Y)
        print(name, score)
```

```python
final_model = models['KNN-3']
```

```python
insurance_validity = input('Insurance Validity: ')
fuel_type = input('Fuel Type: ')
kms_driven = int(input('KMs Driven: '))
ownership = input('ownership:')
transmission = input('Transmission type:')
insurance_validity = mapping[insurance_validity]
fuel_type =mapping1[fuel_type]
ownership =mapping3[ownership]
transmission = mapping2[transmission]
kms_driven = (int)(kms_driven)
test = [[insurance_validity,fuel_type,kms_driven,ownership,transmission]]
yp = final_model.predict(test)[0]
print(f'Predicted Car Price is {yp} Lakh')
```

```python
#create a pickle file
import pickle
pickle.dump(final_model,open('final_model.pkl','wb')) #wb:write binary

#CREATE aa app.py file 

#commands required in anaconda prompt
#1 cd folderpath
#2 streamlit run app.py
```
